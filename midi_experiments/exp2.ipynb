```python
from music21 import midi, stream

def create_midi_stream(notes):
    s = stream.Stream()
    for n in notes:
        s.append(n)
    return s

def export_to_midi(stream, filename):
    mf = midi.translate.music21ObjectToMidiFile(stream)
    for track in mf.tracks:
        for msg in track.events:
            if msg.channel is not None:
                msg.channel = 10  # Canal 10 en términos de índice (0-15)
            if msg.type == 'program_change':
             msg.program = 0  # Establecer el programa en 0 para la percusión
    mf.open(filename, 'wb')
    mf.write()
    mf.close()
```

```python
from music21 import note, instrument

def generate_drum_pattern(pattern):
    drum_kit = instrument.Woodblock()  # Representa un kit de batería
    notes = []
    for beat in pattern:
        if beat == 1:
            n = note.Unpitched("C4")  # Sonido de caja
            n.duration.quarterLength = 0.5
            n.storedInstrument = drum_kit
            notes.append(n)
        else:
            notes.append(note.Rest(quarterLength=0.5))
    return notes
```

```python
def create_drum_loop():
    pattern1 = [1, 0, 1, 0, 1, 0, 1, 0]  # Patrón básico
    pattern2 = [1, 0, 0, 0, 1, 0, 0, 0]  # Variación del patrón
    
    notes = generate_drum_pattern(pattern1) + generate_drum_pattern(pattern2)
    return notes
```

```python
drum_notes = create_drum_loop()
drum_stream = create_midi_stream(drum_notes)
export_to_midi(drum_stream, "drum_loop.mid")
```

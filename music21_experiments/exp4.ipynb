```python
import pygame.midi

# Inicializar pygame.midi
pygame.midi.init()

# Listar los dispositivos MIDI disponibles
for i in range(pygame.midi.get_count()):
    info = pygame.midi.get_device_info(i)
    is_input = info[2]  # True si es dispositivo de entrada (input)
    is_output = info[3]  # True si es dispositivo de salida (output)
    print(f"ID: {i}, Nombre: {info[1].decode()}, Entrada: {is_input}, Salida: {is_output}")
```

```python
import pygame.midi
import time

# Inicializa pygame.midi
pygame.midi.init()

# Selecciona el dispositivo de salida MIDI (reemplaza 0 con el ID que corresponda a tu dispositivo de salida)
output_id = 0  # Reemplazar con el ID del dispositivo de salida que quieres usar
midi_out = pygame.midi.Output(output_id)

# Enviar un mensaje MIDI para reproducir una nota (ejemplo: Nota 60 (Do central) en el canal 1)
midi_out.note_on(60, velocity=127)  # Nota 60 (Do central) con velocidad máxima (127)
time.sleep(1)  # Mantén la nota por 1 segundo
midi_out.note_off(60)  # Apaga la nota

# Cierra el dispositivo MIDI cuando termines
midi_out.close()

# Finaliza pygame.midi
pygame.midi.quit()
```

```python
import pygame

# Inicializar todos los módulos de pygame
pygame.init()

# Inicializar el mixer de pygame para audio/midi
pygame.mixer.init()

# Cargar el archivo MIDI (reemplaza 'ruta/al/archivo.mid' por la ruta de tu archivo MIDI)
pygame.mixer.music.load('C:/Users/Djcho/Documents/Repositories/Polyrhythmia/experiments/DeepGroovySounds.mid')

# Reproducir el archivo MIDI
pygame.mixer.music.play()

# Mantener el programa activo mientras el MIDI se está reproduciendo
while pygame.mixer.music.get_busy():
    pygame.time.Clock().tick(10)  # Espera 10ms para no sobrecargar la CPU

# Finalizar pygame una vez que termine
pygame.quit()
```

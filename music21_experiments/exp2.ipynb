```python
from music21 import converter, midi, instrument, stream, note

snareNote = note.Note(38)
bassFirstNote = note.Note(35, quarterLength=2)
bassSecondNote = note.Note(35, quarterLength=2)
hiHatNote = note.Note(42)

snarePart = stream.Part()
snareDrum = instrument.SnareDrum()
snarePart.insert(snareDrum)
snarePart.insert(snareNote)


bassPart = stream.Part()
bassDrum = instrument.BassDrum()
bassPart.insert(bassDrum)
bassPart.append(bassFirstNote)
bassPart.append(bassSecondNote)

hiHatPart = stream.Part()
hiHat = instrument.HiHatCymbal()
hiHatPart.insert(hiHat)
hiHatPart.repeatAppend(hiHatNote, 4)

score = stream.Score()
score.insert(0, snarePart)
score.insert(0, bassPart)
score.insert(0, hiHatPart)

sp = midi.realtime.StreamPlayer(score)
sp.play()
```

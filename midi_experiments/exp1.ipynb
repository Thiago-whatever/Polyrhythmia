```python
from mido import MidiFile, MidiTrack, Message, MetaMessage

# Crear un nuevo archivo MIDI
mid = MidiFile()

# Crear pista para el piano
piano = MidiTrack()
mid.tracks.append(piano)
piano.append(MetaMessage('set_tempo', tempo=500000))  # 120 BPM
piano.append(Message('program_change', program=0))  # Piano acústico

# Notas para el piano (C - G - Am - F)
notes = [60, 67, 69, 65]  # C, G, A, F (en la octava 5)
durations = [480, 480, 480, 480]  # Duración de las notas en ticks

for note, duration in zip(notes, durations):
    piano.append(Message('note_on', note=note, velocity=64, time=0))
    piano.append(Message('note_off', note=note, velocity=64, time=duration))

# Crear pista para el bajo
bajo = MidiTrack()
mid.tracks.append(bajo)
bajo.append(MetaMessage('set_tempo', tempo=500000))  # 120 BPM
bajo.append(Message('program_change', program=32))  # Bajo eléctrico

# Notas para el bajo (C, G, A, F)
bajo_notes = [36, 43, 45, 41]  # Notas en la octava 2
for note, duration in zip(bajo_notes, durations):
    bajo.append(Message('note_on', note=note, velocity=64, time=0))
    bajo.append(Message('note_off', note=note, velocity=64, time=duration))

# Crear pista para la batería
bateria = MidiTrack()
mid.tracks.append(bateria)
bateria.append(MetaMessage('set_tempo', tempo=500000))  # 120 BPM

# Configuración básica de batería
bateria.append(Message('program_change', program=115))  # Kit de batería

# Notas para la batería
bateria.append(Message('note_on', note=36, velocity=100, time=0))  # Bombo
bateria.append(Message('note_off', note=36, velocity=100, time=240))  # Tiempo de duración

bateria.append(Message('note_on', note=38, velocity=100, time=0))  # Caja
bateria.append(Message('note_off', note=38, velocity=100, time=240))  # Tiempo de duración

bateria.append(Message('note_on', note=42, velocity=100, time=0))  # Hi-Hat
bateria.append(Message('note_off', note=42, velocity=100, time=120))  # Tiempo de duración

# Guardar el archivo MIDI
mid.save('cumbia_del_amor.mid')

print("Archivo MIDI creado con éxito: cumbia_del_amor.mid")
```

```python
from music21 import stream, note, midi, tempo

# Crear un nuevo Stream
s = stream.Stream()

# Ajustar el tempo (bpm - beats per minute)
metronome = tempo.MetronomeMark(number=220)  # tempo de "Billie Jean" es aproximadamente 117 bpm
s.append(metronome)

# Función para crear nuevas notas de batería
def create_note(midi_number, duration):
    n = note.Note()
    n.pitch.midi = midi_number
    n.duration.quarterLength = duration
    return n

# Duración en segundos del patrón de batería
pattern_duration = 1 + 1 + 1 + 1 + 1 + 1 + 1  # Suma de las duraciones de las notas (en beats)

# Calcular cuántos patrones se necesitan para llegar a 30 segundos
total_duration = 30  # Duración deseada en segundos
beats_per_second = metronome.number / 60
patterns_needed = total_duration * beats_per_second / pattern_duration

# Añadir las notas al stream repetidamente hasta alcanzar la duración deseada
for _ in range(int(patterns_needed)):
    s.append(create_note(36, 1))  # Kick Drum
    s.append(create_note(42, 1))  # Hi-Hat
    s.append(create_note(38, 1))  # Snare Drum
    s.append(create_note(42, 1))  # Hi-Hat

# Exportar a archivo MIDI y cambiar el canal a 10 (percusión)
mf = midi.translate.music21ObjectToMidiFile(s)
for track in mf.tracks:
    for msg in track.events:
        if msg.channel is not None:
            msg.channel = 10  # Canal 10 en términos de índice (0-15)
        if msg.type == 'program_change':
            msg.program = 0  # Establecer el programa en 0 para la percusión

mf.open('billie_jean_drum_30s.mid', 'wb')
mf.write()
mf.close()


#Cabe aclarar que una bateria en midi nunca va a sonar igual que la grabación de una batería, por si sola (se necesitaria al menos un programa como Logic Pro, Ableton, etc. , para que los plugins pudieran
#darle el sonido mas cercano a una bateria real); aun asi, se puede escuchar la diferencia entre una grabacion de una bateria y una hecha con la ayuda de la tecnologia, pues no se ha logrado igualar
#el trabajo del baterista, el productor y el mixer para sonidos precargados de una libreria
```
